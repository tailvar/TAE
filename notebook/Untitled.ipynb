XTX intuition

```python
import numpy as np
X=np.array([3,4])
```

```python
norm_sq = np.dot(X, X) # X^T X
norm_sq_alt=np.linalg.norm(X)**2 # ||X||^2
```

```python
print(norm_sq)
print(norm_sq_alt)
```

<x,y> = X^T y intuition

```python
x=np.array([1,2,3])
y=np.array([4,5,6])

print(np.dot(x,y))
print(x.T@y)
```

x^T y = 0 intuition

```python
x=np.array([1,0])
y=np.array([0,1])
print(x.T@y)
```
